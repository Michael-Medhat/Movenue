```python
import ast
```

```python
def split_list(x):
    first = x.find('{')
    second = x.find('}')
    if first == -1 or second == -1:
        return np.nan
    x = x[first:second+1]
    x = ast.literal_eval(x)
    return x['name']
```

```python
temp_df.production_companies = temp_df.production_companies.apply(split_list)
```

using sparse matrix for more efficient computations

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error

from scipy.sparse import csr_matrix
```

```python
X = csr_matrix(final_df.drop(['world_revenue'], axis=1).values)
y = final_df.world_revenue.values
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from xgboost import XGBRFRegressor
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'max_features': ('auto','sqrt'),
             'n_estimators': [500, 1000],
             'max_depth': [10, None],
             'min_samples_split': [5],
             'min_samples_leaf': [1],
             'bootstrap': [True]}
```

```python
grid_search = GridSearchCV(RandomForestRegressor(),parameters,cv=5,return_train_score=True,n_jobs=-1)
grid_search.fit(X,y)
grid_search.best_params_
```

```python
model = RandomForestRegressor(n_jobs=-1,random_state=51 , bootstrap = True,
 max_depth= None,
 max_features = 'auto',
 min_samples_leaf = 7,
 min_samples_split= 5,
 n_estimators = 1000)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train,y_train)
```

```python
mean_squared_error(model.predict(X_train),y_train)**0.5
```

```python
mean_squared_log_error(model.predict(X_train),y_train)**0.5
```

```python
model.score(X_test, y_test)
```

```python
mean_squared_error(model.predict(X_test),y_test)**0.5
```

```python
mean_squared_log_error(model.predict(X_test),y_test)**0.5
```
